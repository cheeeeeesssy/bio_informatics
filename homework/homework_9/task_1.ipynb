**1 задание**  
Сгенирируйте в python равномерное распределение и постройте для него ящик с усами  
Для следующих данных:  
79  80  80  80  74  80  80  79  64  78  73  78  74  45  81  48  80  82  82  70  
(a.) Получите квартили.   
(b.) Определите межквартильный размах.   
(c.) Найдите минимум, Q25, медиану, Q75 и максимум   
(d.) Определите потенциальные выбросы, если таковые имеются.  
(e.) Постройте ящик с усами

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
import seaborn as sns
```

```python
uniform_data = sps.uniform.rvs(size=100)
plt.figure(figsize=(10, 7))
sns.boxplot(x=uniform_data, color='skyblue')
plt.title('Boxplot для равномерно распределённых данных')
plt.xlabel('Значения')
plt.grid(True)
plt.show()
```

```python
data = np.array([79, 80, 80, 80, 74, 80, 80, 79, 64, 78, 73, 78, 74, 45, 81, 48, 80, 82, 82, 70])
q1 = np.percentile(data, 25)
q2 = np.median(data)
q3 = np.percentile(data, 75)
print(f"Q1(25%) = {q1}")
print(f"Q2(50%) = {q2}")
print(f"Q3(75%) = {q3}")
```

```python
iqr = q3 - q1
print(f"Межквартильный размах: {iqr}")
```

```python
print(f"Минимум: {data.min()}")
print(f"Q25: {q1}")
print(f"Медиана: {q2}")
print(f"Q75: {q3}")
print(f"Максимум: {data.max()}")
```

```python
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
outliers = data[(data < lower) | (data > upper)]
print("Выбросы:", outliers)
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=data, color='skyblue')
plt.title('Boxplot для равномерно распределённых данных')
plt.xlabel('Значения')
plt.grid(True)
plt.show()
```
